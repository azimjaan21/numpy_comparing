# NumPy Basic Array Methods




1. **np.zeros**

```python
# you can create multidimensional arrays of zeros by passing
# a tuple with the 1st value being rows and the 2nd columns

#arr1 = np.zeros(3)
#arr1

arr = np.zeros((3,3))  # it has 3 rows and 3 columns
arr
```

2. **np.ones**

```python
# Create array of 1s

arr1 = np.ones((4,3))
arr1
```

3. **np.arange**

```python
# Creates an array with a range of values.

arr = np.arange(15) # [0, 15)
arr

arr_manual = np.arange(0, 15, 4) #3rd is STEP( it adds every time when array's element is adding)
arr_manual
```

4. **np.size**

```python
# to clarifying count of elements in array
arr = np.ones((3,7))
arr_size = np.size(arr) # 3x7 --> 21 elements
arr_size
```

5. **np.linspace**

```python
#Creates an array with a specified number of evenly spaced values between a start and end value.
arr = np.linspace(0, 1, 5)
arr
```

6. **np.reshape**

```python
#Reshapes an array to a specified shape.
arr = np.array([1, 2, 3, 4, 5, 6])
arr_reshaped = np.reshape(arr, (3, 2))
arr_reshaped
```

7. **np.sum**

```python
arr = np.array([1, 2, 3, 4, 5])
arr_sum = np.sum(arr) #Computes the sum of array elements along a specified axis.
arr_sum
```

8. **np.min / np.max**

```python
arr = np.array([1, 2, 3, -4, 5])
arr_min = np.min(arr) #Finds the minimum value in an array.
arr_min
```

9. **np.sort**

```python
arr = np.array([5, 2, 8, 1, 7, 0, -21])
arr_sorted = np.sort(arr) #Sorts the elements of an array.
arr_sorted
```

10. **np.unique**

```python
arr = np.array([1, 2, 2, 3, 4, 4, 5])
arr_unique = np.unique(arr) #Finds the unique elements in an array.
arr_unique
```
